```python
# First, import the relevant modules
import requests
import os
import collections
import json
```

```python
# Now, call the Quandl API and pull out a small sample of the data (only one day) to get a glimpse
# into the JSON structure that will be returned
URL_AFX_X = 'https://www.quandl.com/api/v3/datasets/FSE/AFX_X/data.json'
r = requests.get(URL_AFX_X+'?api_key='+API_KEY+'&start_date=2017-01-02&end_date=2017-01-02')
```

```python
#Transform into dictionary
file=r.json()
type(file)
```

```python
# Inspect the JSON structure of the object you created, and take note of how nested it is,
# as well as the overall structure

#Exploring initial structure - looking into the nested dictionaries
n=0
p=0
for key in file.keys():
    #Print the main keys
    n=n+1
    print(n, "---", key)
    
    #Unpack subkeys
    for subkey in file[key].keys():
        p=p+1
        print(n,'.',p,":",subkey)
        
        #print values
        print(file[key][subkey])
        print("\n")
```

```python
# Task 1: collect data for entire 2017 year 
URL = 'https://www.quandl.com/api/v3/datasets/FSE/AFX_X/data.json'
r= requests.get(URL+'?api_key='+API_KEY+'&start_date=2017-01-01&end_date=2017-12-31')
# Task 2: Convert to JSON
AFX=r.json()
```

```python
cols=AFX['dataset_data']['column_names']
data=AFX['dataset_data']['data']
```

**Calculate what the highest and lowest opening prices were for the stock in this period.**

```python
opening=[i[1] for i in data]
op = list(filter(None, opening)) 

print('Minimum opening:', min(op))
print('Maximum opening:', max(op))
```

**What was the largest change between any two days (based on Closing Price)?**

**What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)**

```python
def median(input):
    input=sorted(input)
    term=int(len(input)/2)-1
    
    if len(input)%2!=0:
        median=input[term]
    else:
        median=(input[term]+input[term+1])/2
    
    print('Median is:', median)
```
