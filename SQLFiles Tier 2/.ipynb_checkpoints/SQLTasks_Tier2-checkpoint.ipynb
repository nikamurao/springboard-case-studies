```python
import pandas as pd
```

```python
from sqlalchemy import create_engine
```

```python
engine=create_engine('sqlite:///sqlite_db_pythonsqlite.db')
```

```python
engine.table_names()
```

```python
#Test - getting the different tables for easy access
with engine.connect() as conn:
    b= conn.execute('SELECT * FROM Bookings')
    bookings=pd.DataFrame(b.fetchall())
    bookings.columns=b.keys()
    
    f= conn.execute('SELECT * FROM Facilities')
    facilities=pd.DataFrame(f.fetchall())
    facilities.columns=f.keys()
    
    m= conn.execute('SELECT * FROM Members')
    members=pd.DataFrame(m.fetchall())
    members.columns=m.keys()
    
```

Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members!

```python
#Initial answer
#query1='SELECT name, total.revenue FROM (SELECT name, SUM( membercost * slots ) AS revenue FROM Facilities JOIN Bookings USING ( facid ) WHERE memid !=0 GROUP BY name) AS total WHERE total.revenue <1000 ORDER BY total.revenue'
```

```python
#Suggested answer:
query1='SELECT name, revenue FROM (SELECT name, SUM(CASE WHEN memid = 0 THEN guestcost * slots ELSE membercost * slots END) AS revenue FROM cd.bookings INNER JOIN cd.facilities ON cd.bookings.facid = cd.facilities.facid GROUP BY name) AS inner_table WHERE revenue < 1000 ORDER BY revenue'
```

```python
with engine.connect() as conn:
    q1=conn.execute(query1)
    q1answer=pd.DataFrame(q1.fetchall())
    q1answer.columns=q1.keys()
```

```python
q1answer
```

```python
#Suggested answer
query1='SELECT f.name,concat(m.firstname,' ',m.surname) as Member, count(f.name) as bookings FROM Members m inner join Bookings bk on bk.memid = m.memid inner join Facilities f on f.facid = bk.facid where m.memid>0 group by f.name,concat(m.firstname,' ',m.surname) order by f.name,m.surname,m.firstname'
```

```python
with engine.connect() as conn:
    q1=conn.execute(query1)
    q1answer=pd.DataFrame(q1.fetchall())
    q1answer.columns=q1.keys()
```

 Q12: Find the facilities with their usage by member, but not guests

```python
query2="SELECT name, memid, SUM( slots ) /2 AS total_hours FROM Bookings JOIN Facilities USING ( facid )  WHERE memid !=0 GROUP BY name, memid"
```

```python
with engine.connect() as conn:
    q2=conn.execute(query2)
    q2answer=pd.DataFrame(q2.fetchall())
    q2answer.columns=q2.keys()
```

```python
q2answer
```

Q13: Find the facilities usage by month, but not guests

```python
query3="SELECT name, strftime('%m', starttime) AS month_, SUM(slots)/2 AS total_hours FROM Bookings JOIN Facilities USING (facid) WHERE memid !=0 GROUP BY name, month_"
```

```python
with engine.connect() as conn:
    q3=conn.execute(query3)
    q3answer=pd.DataFrame(q3.fetchall())
    q3answer.columns=q3.keys()
```
